# Forecast and Inventory Visualization

This notebook visualizes model performance and inventory recommendations.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

fig_dir = Path('../reports/figures')
fig_dir.mkdir(parents=True, exist_ok=True)
```

```python
forecast = pd.read_csv('../reports/forecast_actual_vs_predicted.csv')
forecast['WeekStart'] = pd.to_datetime(forecast['WeekStart'])
forecast.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(forecast['WeekStart'], forecast['actual'], label='Actual')
plt.plot(forecast['WeekStart'], forecast['prediction'], label='Predicted')
plt.legend()
plt.title('Demand Forecast vs Actual')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(forecast['WeekStart'], forecast['actual'], label='Actual')
plt.plot(forecast['WeekStart'], forecast['prediction'], label='Predicted')
plt.legend()
plt.title('Demand Forecast vs Actual')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(fig_dir / 'forecast_plot.png', dpi=160)
plt.show()
```

```python
inventory = pd.read_csv('../reports/inventory_recommendations.csv')
inventory_plot = inventory.rename(columns={'StockCode': 'product_id', 'ReorderPointQty': 'reorder_point'})
inventory_plot.head()
```

```python
inventory_plot.sort_values('reorder_point', ascending=False).head(10).plot(
    x='product_id',
    y='reorder_point',
    kind='bar',
    figsize=(10,5),
)
plt.title('Top Products by Reorder Point')
plt.tight_layout()
plt.show()
```

```python
ax = inventory_plot.sort_values('reorder_point', ascending=False).head(10).plot(
    x='product_id',
    y='reorder_point',
    kind='bar',
    figsize=(10,5),
)
ax.set_title('Top Products by Reorder Point')
plt.tight_layout()
plt.savefig(fig_dir / 'inventory_reorder_top10.png', dpi=160)
plt.show()
```
